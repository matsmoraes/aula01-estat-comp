Gráficos

```python
plt.hist(df['Palavras.sentença'], 5, rwidth=0.99) # (Coluna do dataframe, numero de colunas, espaçamento entra as colunas)
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize= (10, 6))
df['Palavras.sentença'].hist(bins = 5, rwidth = 0.99, color = 'royalblue')
plt.title("Histograma")
plt.xlabel("Palavras por sentença")
plt.ylabel("Frequência")
plt.show()
```

```python
# Gráfico de pontos
x = df['Palavras.sentença']
plt.plot(x, 'o')
plt.xticks(np.arange(len(x)), np.arange(1,len(x)+1))
#plt.xtricks(local no eixo, marca/nome)
plt.xlabel("Página")
plt.ylabel("Palavras por sentença")
plt.show()
```

```python
# Gráfico de linhas
x = df['Palavras.sentença']
plt.plot(x)
plt.xticks(np.arange(len(x)), np.arange(1,len(x)+1))
#plt.xtricks(local no eixo, marca/nome)
plt.xlabel("Página")
plt.ylabel("Palavras por sentença")
plt.show()
```

```python
# Gráfico de dispersão
x = df['Palavras.sentença']
y = df['Legibilidade']
plt.plot(x,y, 'o')
#plt.xticks(np.arange(len(x)), np.arange(1,len(x)+1))
#plt.xtricks(local no eixo, marca/nome)
plt.xlabel("Página")
plt.ylabel("Legibilidade")
plt.show()
```

```python
# Gráfico de dispersão (outra forma)
x = df['Palavras.sentença']
y = df['Legibilidade']
plt.plot(x,y, 'o')
plt.scatter(x,y,c='b',s=30) # s permite mudar o 'tamanho'
plt.xlabel("Página")
plt.ylabel("Legibilidade")
plt.show()
```
